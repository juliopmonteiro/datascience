# Pré-processamento dos dados

## Transformando colunas categóricas em numéricas

É necessário realizar uma conversão do tipo de dado, pois modelos de ML não suportam dados que não sejam numéricos.

```python
employee_df.head()
```

```python
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

Quando temos colunas que suas variáveis não apresentam um nível de importância entre elas, como o nível de escolaridade, para defini-las usamos variáveis Dummy, que servem como uma espécie de "tabela verdade", segue exemplo:

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

## DF com as variáveis numéricas

```python
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
```

## Unindo os dois DF (contatenando)

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
X
```

## Criando variável Y

Na variável X, atribuímos os os atributos previsores, aqueles que irão ser analizados, e devo criar uma outra variável com o atributo a ser analisado (y), nesse caso, o Attrition! 

```python
y = employee_df['Attrition']
y
```

# Implementação Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)     #Fit encaixa os dados de treinamento na minha base
```

```python
# Aplicando para que o Algoritmo faça previsões
y_pred = logistic.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

Nosso modelo tem uma Accuracy Score de 87%, o que é uma boa porcentagem, porém é necessário analisar o F1-SCORE (Precision x Recall ), para ver a precisão ao definir se um funcionário vai sair da empresa, o que é o objetivo da empresa (estudo de caso)!

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True);
```

Apresenta uma dificuldade em prever funcionários que sairão da empresa, com um alto número de falsos negativos (40).

```python
# Precision = TP/(TP+FP)
20/(20+5)
```

```python
# Recall = TP/(TP+FN)
20/(20+40)
```

Agora com a teoria fixada, vamos realizar os mesmo cálculos de maneira mais prática, usando o Sklearn ;)

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='macro')
# O 'average=macro' irá pegar o F1score para cada uma das possibilidades de saída e fazer uma média entre elas
```

```python
print(classification_report(y_test, y_pred))
```

# Implementação Redes Neurais

- Quantos neurônios e quantas camadas? https://iaexpert.academy/2020/05/04/quantas-camadas-escondidas-e-quantos-neuronios-incluir-numa-rede-neural-artificial/

```python
import tensorflow as tf
```

```python
X_train.shape
```

```python
# Para saber quantos neurônios colocar na minha Rede Neural, posso pegar o número de atributos (colunas), somar com o número de saídas (1) e dividir por 2!
(50 + 1) / 2
```

```python
rede_neural = tf.keras.models.Sequential()
rede_neural.add(tf.keras.layers.Dense(units = 25, activation='relu', input_shape=(50,))) #Input shape vai ser igual ao número de atributos/colunas
rede_neural.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
rede_neural.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))  #Basicamente, a função de ativação 'relu' zera as saídas negativas e retorna igual para as positivas
rede_neural.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
rede_neural.summary()
```

Teremos 2601 pesos (w) a serem atualizados. O objetivo da rede neural é encontrar o melhor conjunto de pesos.

```python
rede_neural.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy']) 
```

```python
# Treinamento da Rede Neural
rede_neural.fit(X_train, y_train, epochs=200)
```

```python
y_pred = rede_neural.predict(X_test)
y_pred
```

```python
y_pred = (y_pred >= 0.5)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True);
```

```python
print(classification_report(y_test, y_pred))
```
